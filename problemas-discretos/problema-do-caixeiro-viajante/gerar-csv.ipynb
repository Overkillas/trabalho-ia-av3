```python
import numpy as np
import pandas as pd

N_PONTOS = 50
LIMITE_MIN = -30
LIMITE_MAX =  30 
NOME_ARQUIVO = "CaixeiroGrupos.csv"

np.random.seed(42)

pontos = np.random.uniform(LIMITE_MIN, LIMITE_MAX, size=(N_PONTOS, 3))

df = pd.DataFrame(pontos, columns=["x", "y", "z"])
df.to_csv(NOME_ARQUIVO, index=False)

print(f"\nArquivo CSV '{NOME_ARQUIVO}' gerado com sucesso!")
print(f"Total de pontos: {N_PONTOS}")
print(f"Limites: {LIMITE_MIN} a {LIMITE_MAX}\n")
```
